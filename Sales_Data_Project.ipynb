```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv('sales_data.csv',encoding='latin1')
ds
```

```python
ds.info()
```

```python
ds=ds.drop(['Status','unnamed1'],axis=1)
ds
```

```python
ds=ds.reset_index()
```

```python
ds.drop('index',axis=1,inplace=True)
ds
```

```python
ds.dropna(inplace=True)
ds
```

```python
ds.rename(columns={'Marital_Status':'Marital'},inplace=True)
ds
```

```python
ds['Marital']=ds['Marital'].map({0:'single',1:'married'})
ds
```

```python
ds['Gender']=ds['Gender'].map({'F':'Female','M':'Male'})
ds
```

```python
fig=plt.figure(figsize=(10,10),facecolor='tan')
plt.gcf().suptitle('Customer Distribution',fontsize=17)

plt.subplot(2,2,1)
g_per=ds['Gender'].value_counts()
plt.pie(g_per.values,labels=g_per.index,autopct='%0.2f%%',radius=1.2,pctdistance=0.5,colors=['bisque','saddlebrown'],explode=[0,0.08],textprops={'color':'k'},startangle=90,wedgeprops={'edgecolor':'sienna','linewidth':1.5})
plt.title('Gender Disturibution of Customers')


plt.subplot(2,2,2)
count=sns.countplot(y='Gender',data=ds,hue='Gender',palette='gist_heat',width=0.8)
for i in count.containers:
  count.bar_label(i,fontsize=10,color='snow',rotation=0,label_type='center')
plt.title('Customer Count by Gender')
plt.gca().set_facecolor('ivory')


plt.subplot(2,2,3)
count=sns.countplot(y='Marital',data=ds,hue='Marital',palette='autumn',width=0.8)
for i in count.containers:
  count.bar_label(i,fontsize=10,color='snow',rotation=0,label_type='center',fmt='%d')
plt.title('Marital Status Distribution')
plt.gca().set_facecolor('ivory')

plt.subplot(2,2,4)
g_per=ds['Marital'].value_counts()
plt.pie(g_per.values,labels=g_per.index,autopct='%0.2f%%',radius=1.2,pctdistance=0.8,colors=['beige','chocolate'],shadow=True,textprops={'color':'k'},startangle=90,wedgeprops={'edgecolor':'sienna','linewidth':1.5,'width':0.5})
plt.title('Customer Count by Marital Status')

plt.tight_layout()
fig.savefig('sales_data1.jpg',dpi=500)
plt.show()

```

**Conclusion**

The sales data indicates that the customer base is predominantly **female**, accounting for nearly **70% of total customers (7,832)**, while **male customers represent about 30% (3,407)**. This highlights a strong female-driven market segment, suggesting that products, promotions, and communication strategies aligned with female preferences are likely to yield higher engagement and conversions.

From a marital status perspective, **single customers form the majority at approximately 58% (6,518)**, compared to **married customers at 42% (4,721)**. This suggests that purchasing behavior is more influenced by individual consumers rather than family-oriented buyers.

Overall, the sales data implies that targeting **single female customers** should be a primary focus for marketing and product positioning. However, the sizable presence of married customers also presents opportunities for tailored family or household-oriented offerings to further expand sales potential.

**Conclusion**

The sales analysis reveals clear patterns across demographics, geography, occupation, and product categories. **Female customers generate significantly higher sales than male customers**, reinforcing the importance of women as the primary revenue-driving segment. In terms of age, the **26â35 age group contributes the highest sales**, followed by **36â45 and 18â25**, indicating that young and mid-career adults form the core purchasing audience.

Occupation-wise, the **IT sector, Healthcare, and Aviation** emerge as the top contributors to sales, suggesting strong purchasing power among professionals in these industries. This insight can help tailor industry-specific promotions and partnerships.

Geographically, **Uttar Pradesh, Maharashtra, and Karnataka** lead in sales volume, making them key markets for expansion, localized campaigns, and distribution optimization. Other states such as Delhi and Andhra Pradesh also show solid performance, indicating multi-regional demand.

From a marital status perspective, **single customers contribute more to total sales than married customers**, aligning with earlier customer distribution trends and suggesting higher discretionary spending among single individuals.

Regarding product categories, **Food, Clothing & Apparel, Electronics & Gadgets, and Footwear & Shoes** dominate sales, showing strong demand for daily-use and lifestyle products. Categories like Office supplies, Veterinary, and Pet Care generate relatively lower revenue, representing potential areas for targeted growth strategies.

Overall, the business should prioritize **female customers aged 18â45, professionals in IT and Healthcare, and high-performing states**, while continuing to invest in top-selling product categories and selectively boosting underperforming segments to achieve balanced growth.
